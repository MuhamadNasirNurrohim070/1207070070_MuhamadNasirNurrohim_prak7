```python
import cv2 #mengimport atau memanggil cv2 dari opencv untuk pengolahan citra serta pembacaan gambar
import numpy as np #import library numpy untuk operasi numerik
from skimage import data #import data dari skimage sbagai acuan atau contoh 
import matplotlib.pyplot as plt #import matplotlib yang digunakan sebagai visualisasi gambar
#matplotlib inline
```

```python
img = cv2.imread("gambar/cyber.jpg", cv2.IMREAD_GRAYSCALE)  #membaca atau menampilkan gambar 

row, column = img.shape  #berfungsi untuk mendapatkan dimensi citra

img1 = np.zeros((row, column), dtype='uint8')  #pembuatan array kosong dengan data yang sama

min_range = 10  #nilai interval range bawah
max_range = 60  #nilai interval range atas

# proses pengubahan pixel menggunakan if else berdasarkan interval range 
for i in range(row):
    for j in range(column):
        if img[i, j] > min_range and img[i, j] < max_range:
            img1[i, j] = 255
        else:
            img1[i, j] = 0
```

```python
#pembuatan subplot untuk menampilkan hasil dari gray level slicing
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title("Citra Input")
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```
